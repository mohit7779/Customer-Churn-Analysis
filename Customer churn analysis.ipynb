```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
    
```

### converted 0 and 1 value of senior citizen to yes/no to make it easier to understand.

# churn

```python
ax = sns.countplot(x = "Churn",data = df,hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customer By Churn")
# Q how much customer have churned out?
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%")
plt.title("percentage of Customer By Churn")
# Q how much % of customer have churned out?
```

## from the given pie chart we can conclude that 26.54% of customer have churned Out.

```python
sns.countplot(x = "gender",data  = df,hue = "Churn")
plt.title("Churn By Gender",fontsize = 20)
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure",data  = df,bins = 72,hue = "Churn")
```

##  people who have used our service for a long time have stayed and people who have used our services # 1 or month have churned

```python
ax = sns.countplot(x = "Contract",data  = df,hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customer By Contract",fontsize = 15)
```

## people who have month to month contract are likely to Churn than from those who have 1 or 2 Year of Contract

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 
           'StreamingTV', 'StreamingMovies']

# Set up the plot grid
plt.figure(figsize=(12, 20))  # Adjust the figure size as needed

# Loop through each column and create a count plot
for i, col in enumerate(columns, 1):
    plt.subplot(5, 2, i)  # Create a subplot for each variable (5 rows, 2 columns)
    sns.countplot(x=col, data=df,hue = "Churn")
    plt.title(f'Count plot of {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed

plt.tight_layout()  # Adjust the layout so the plots don’t overlap
plt.show()
```

## a majority of customers do not churn, as indicated by the higher number of "No" bars across categories. InternetService and related services like OnlineSecurity and StreamingTV have more balanced distributions between customers who churn and those who don’t. 

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x = "PaymentMethod",data = df,hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customer By Payment Method",fontsize = 15)
```
